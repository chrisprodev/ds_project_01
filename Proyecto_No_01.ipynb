# Ejercicio N°1:

Escribir un programa que pregunte al usuario o genere las ventas
de los últimos 5 días, y muestre por pantalla una serie (Dataframe)
con los datos de las ventas por producto (opcional): Nro de
unidades y precio unitario, indexada por los días, antes y después
de aplicarles un descuento del 5%.

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(10)

# creando la data en forma de diccionario
data = {
    "Day":['Monday','Tuesday','Wednesday','Thursday','Friday'],
    "Product":['Hula Hoop', 'Chia Pet', 'Tickle Me Elmo', 'Furby', 'Baseball bat'],
    "Qty":np.random.randint(1, 20, 5),
    "Price":np.random.randint(1000, 10000, 5)/100,
}

# creando DataFrame
dataFrame = pd.DataFrame(data)

# agregando columna discount
dataFrame['Discount(5%)'] = (dataFrame['Price'] * .95).round(2)

# agregando columna Total Sales
dataFrame['Total Sales'] = (dataFrame['Qty'] * dataFrame['Discount(5%)'])

# seteando la columna Days como indices
dataFrame = dataFrame.set_index(['Day'])
print(dataFrame)
```

# Ejercicio No 3

Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la siguiente tabla:

```python
# creando el DataFrame a utilizar
sales_df = pd.DataFrame(data={
    "Mes":['Enero','Febrero','Marzo','Abril','Mayo'],
    "Ventas":[30500,35600,28300,33900,42500],
    "Gastos":[22000,23450,18100, 35700,32450]
})
print(sales_df)
```

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
# agregando Columna 1
sales_df['Columna 1'] = sales_df['Gastos']/sales_df['Ventas']*100

# agregando Columna 1
sales_df['Columna 2'] = sales_df['Ventas'].apply(lambda x: 'Meta Superada' if x >=30000 else 'Meta no Superada') 

# agregando Columna 3
sales_df.loc[sales_df['Gastos'] > sales_df['Ventas'], 'Sobre gasto'] = sales_df['Gastos']-sales_df['Ventas']

print(sales_df, '\n')
# Total ventas últimos 3 meses
df_ventas = sales_df.iloc[-3:, 1:2 ]
# Total gastos febrero a abril
df_gastos = sales_df.iloc[1:4, 2:3]

print(df_ventas.sum())
print(df_gastos.sum())
```
